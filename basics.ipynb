```python
2. Dictionaries
```

```python
books = []

book = { 'title': 'The Alchemist',
          'author': "Paulo Coelho",
          'year': 2000
        }
books.append(book)

book = { 'title': 'Harry Potter',
          'author': 'JK Rowling',
          'year': 2002
        }
books.append(book)

book = { 'title': 'Siddhartha',
          'author': 'Hermann Hesse',
          'year': 1922
        }
books.append(book)

print(books)

for book in books:
    name = book['title']
    print(name)
```

```python
3. Functions
```

```python
num = int(input("Enter an integer: "))
def is_even(num):
    # Check if the number is divisible by 2
    if num % 2 == 0:
        return True
    else:
        return False
print(is_even(num))
```

```python
def calculate_area(radius):
    # use 3.14 for pi and calculate the area
    pi = 3.14
    area = pi * (radius ** 2)
    return area
radius = int(input("Enter the radius of the circle:"))
print(calculate_area(radius))
```

```python
3. Classes
```

```python

class Person:
    def __init__(self, name, age, gender):
        # Initialize the attributes
        self.name = name
        self.age = age
        self.gender = gender

    def introduce(self):
        # Print an introduction statement
        print(f"Hello, my name is {self.name}, I am {self.age} years old.")
        
# Create an instance of Person
person = Person("Ram", 25, "Male")

# Call the introduce method
person.introduce()
```
